### <span style="color:Grey">2. Título tentativo del proyecto<span>

"Analisis del costo atención en los pacientes con multimorbilidad según envejecen los pacientes?"

###  <span style="color:Grey">3. Declaración del problema desde el usuario<span>

Por desgracia, las personas enfermas, tienden a enfermarse más, en especial cuando son enfermedades crónicas. Esto provoca muchas veces que el tiempo de hospitalización, tratamiento y costos sea mayor al presupuestado sin los cuidados adecuados.

Por tanto, es de interés de la salud pública cuales son los aumentos de costos en los gastos publicos en referencia a la edad, para así poder generar planes preventivos de manera eficiente sin utilizar la escasa cantidad de recursos en sobreestudiar pacientes que no presentarán gravedad hasta años después, ni llegar tarde a tratar preventivamente.

###  <span style="color:Grey">4. Pregunta principal de investigación<span>

Como se altera el comportamiento de los costos de atencion segun la edad y enfermedades de los pacientes con murtimorbilidad?

###  <span style="color:Grey">5. Preguntas secundarias<span>

- **Pregunta secundaria 1:** ¿Existe una relación significativa entre el costo y la edad de los pacientes?
- **Pregunta secundaria 2:** ¿Como distribuyen las personas con multimorbilidad en relación al costo?
- **Pregunta secundaria 3:** ¿Existen regiones cuyos costos sean diferentes en comparación al país?

###  <span style="color:Grey">6. Variables principales y justificación de selección<span>

- **Variable 1:** Edad - Es una variable independiente que nos indica la edad del paciente
- **Variable 2:** Región - Es una variable dependiente de la varible centro de salud, la cual nos indica la región en la que se atendio el paciente
- **Variable 3:** Peso - Es una variable independiente la cual nos indica el costo del tratamiento
- **Variable 4:** Tipo Ingreso - Variable que indica si el paciente entro por urgencia o por procedimiento
- **Variable 5:** Dias Estadia - Variable dependiente que se cacula con la fecha de ingreso y de salida, esta nos indica el tiempo de recuperación o desceso del paciente
- **Variable 6:** Severidad - Variable independiente que indica el nivel de severidad del paciente
- **Variable 7:** Diagnostico - Variable que indica la enfermedad que posee el paciente

###  <span style="color:Grey">7. Metodología estadística propuesta<span>

- **Método 1:** Análisis de varianza (ANOVA) - Comparar los costos de atención médica entre diferentes grupos de edad y niveles de multimorbilidad.
- **Método 2:** Análisis de varianza (ANOVA) - Evaluar si existen diferencias significativas en los costos según la cantidad de enfermedades crónicas que presentan los pacientes.
- **Método 3:** Regresión lineal múltiple - Modelar la relación entre la edad, el número de enfermedades crónicas y los costos de atención médica.
- **Método 4:** Regresión lineal múltiple - Incluir variables de control como el sexo, fecha ingreso y el servicio de salud para analizar su impacto en los costos.
- **Método 5:** Pruebas de ajuste de distribución - Identificar distribuciones estadísticas que mejor se ajusten a los costos medicos.
- **Método 6:** Análisis de Clustering o Agrupamiento de Pacientes - Identificar grupos de pacientes con características similares en cuanto a edad, comorbilidades y el costo de atención médica, utilizando técnicas de agrupamiento.
- **Método 7:** Análisis de Costos Incrementales - Comparar los costos adicionales de atención médica debido a la presencia de comorbilidades en comparación con pacientes sin comorbilidades.

###  <span style="color:Grey">9. Supuestos estadísticos necesarios<span>

- **Supuesto 1:** Normalidad de los datos - Validación con pruebas de Shapiro-Wilk y gráficos Q-Q.
- **Supuesto 2:** Homocedasticidad - Validación con prueba de Breusch-Pagan y gráficos de residuos.
- **Supuesto 3:** Independencia de observaciones - Validación mediante análisis de autocorrelación y prueba de Durbin-Watson.
- **Supuesto 4:** Linealidad - Validación mediante graficos de dispercion y pruebas de linealidad.
- **Supuesto 5:** Multicolinealidad - Validación mediante el Índice de Condición o el VIF.
- **Supuesto 6:** Equivalencia de grupos - Validación a través de análisis de covarianza (ANCOVA) o pruebas de balance de covariables.
- **Supuesto 7:** Normalidad de los costos médicos - Validación con pruebas de Kolmogorov-Smirnov, Anderson-Darling y/o Shapiro-Wilk.
